```python
import requests
import pandas as pd
import credentials
url = "https://spotify23.p.rapidapi.com/search/"
headers = credentials.Albums_header
```

```python
# getting albums data
getalbums=pd.DataFrame()
for i in lower:
    querystring = {"q":i,"type":"albums","offset":"0","limit":"50","numberOfTopResults":"5"}
    response = requests.get(url, headers=headers, params=querystring)
    data=response.json()
    data1=pd.json_normalize(data,['albums',['items']])
    getalbums=pd.concat([getalbums,data1])
getalbums
```

```python
artists_uri=[getalbums['data.artists.items'].values[i][0]['uri'] for i in range(len(getalbums))]
artists_uri
```

```python
albums=pd.DataFrame()
albums_exploded=getalbums.explode(['data.artists.items'])
# albums_exploded['test'] = json.loads(json.dumps(str(albums_exploded['data.artists.items']).replace("0",'').strip()))
albums['album_uri']=albums_exploded['data.uri']
albums['album_name']=albums_exploded['data.name']
albums['album_artist_uri']=[i['uri'] for i in albums_exploded['data.artists.items']]
albums['artist_name']=[i['profile']['name'] for i in albums_exploded['data.artists.items']]
albums['release_year']=albums_exploded['data.date.year']
albums
# import json
# t = {'uri': 'spotify:artist:6vUn7EgNei8hY7jZSV0lsT',
#         'profile': {'name': 'a picture of her'}}
# t['uri']
```

```python
albums['album_uri']=[i[14::] for i in albums['album_uri']]
albums['album_artist_uri']=[i[15::] for i in albums['album_artist_uri']]
albums.to_csv('album.csv',index=False)

```
