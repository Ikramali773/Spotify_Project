```python
import pandas as pd
albumsdf=pd.read_csv('Albums.csv')
albumsdf
```

```python
artistid=list(albumsdf['artist_id'])
artistid
```

```python
import requests
import credentials
url = "https://spotify23.p.rapidapi.com/artist_discography_overview/"
headers = credentials.Artist_header
artists_overview=pd.DataFrame()
for i in artistid:
    querystring = {"id":i}
    response = requests.get(url, headers=headers, params=querystring)
    data= response.json()
    data1=pd.json_normalize(data)
    artists_overview=pd.concat([artists_overview, data1])
```

```python
artists_overview
```

```python
artists_overview.to_csv('artist_discography.csv')
```

```python
import requests
import pandas as pd
import credentials
url = "https://spotify23.p.rapidapi.com/artists/"

headers = credentials.Artist_header

all_artists_df= pd.DataFrame()

for i in artistid:
    querystring = {"ids":i}
    response = requests.get(url, headers=headers, params=querystring)
    data=response.json()
    data1=pd.json_normalize(data['artists'])
    all_artists_df=pd.concat([all_artists_df, data1])
```

```python
all_artists_df
```

```python
all_artists_df.to_csv('artist.csv')
```
